```python
import pandas as pd
import random
from faker import Faker

fake = Faker()
random.seed(42)

data = []

for _ in range(10000):  # 10K records
    student_id = fake.uuid4()
    age = random.randint(17, 25)
    gender = random.choice(['Male', 'Female', 'Other'])
    gpa = round(random.uniform(1.0, 4.0), 2)
    attendance_rate = random.randint(40, 100)
    missed_assignments = random.randint(0, 10)
    library_visits = random.randint(0, 30)
    lms_logins_per_week = random.randint(0, 20)
    screen_time_hours = round(random.uniform(3, 12), 1)
    sleep_hours = round(random.uniform(3, 9), 1)
    study_hours_per_day = round(random.uniform(0, 6), 1)
    financial_stress_score = random.randint(1, 10)
    help_requests = random.randint(0, 5)
    stress_score = random.randint(1, 10)

    # Rule-based risk assignment
    if (gpa < 2.3 and attendance_rate < 60) or (stress_score > 8 and sleep_hours < 5):
        mental_health_risk = 'High'
    elif (stress_score > 6 or missed_assignments > 5 or lms_logins_per_week < 4):
        mental_health_risk = 'Medium'
    else:
        mental_health_risk = 'Low'

    data.append({
        'student_id': student_id,
        'age': age,
        'gender': gender,
        'gpa': gpa,
        'attendance_rate': attendance_rate,
        'missed_assignments': missed_assignments,
        'library_visits': library_visits,
        'lms_logins_per_week': lms_logins_per_week,
        'screen_time_hours': screen_time_hours,
        'sleep_hours': sleep_hours,
        'study_hours_per_day': study_hours_per_day,
        'financial_stress_score': financial_stress_score,
        'help_requests': help_requests,
        'stress_score': stress_score,
        'mental_health_risk': mental_health_risk
    })

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("student_mental_health_data.csv", index=False)

print("✅ Synthetic dataset created with shape:", df.shape)
df.head()
```
